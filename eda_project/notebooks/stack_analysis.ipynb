```python
import yfinance as yf
import pandas as pd
import plotly.express as px

# Define the Tickers
# AAPL = Tech (Growth), WMT = Walmart (Defensive), SPY = S&P 500 (Benchmark)
tickers = ["AAPL", "WMT", "SPY"]

# Import the data
data = yf.download(tickers, start="2022-01-01", end="2024-01-01", auto_adjust=True)["Close"]

# Calculate the daily returns
returns = data.pct_change().dropna()

display(data.head())
display(returns.head())
```

```python
# Plot the price of AAPL, SPY, and WMT over the course of 2022-2024
fig_price = px.line(data, title="Stock Prices (AAPL, SPY , WMT)", labels= {"value": "Stock Price ($)", "Date": "Month Year"})
fig_price.show()
```

From the figure, we can see that the stock price of WMT is the lowest around $50 and SPY has the highest around $400+. Next, we will normailze the data by calculating cumuative returns to compare stocks fairly.

```python
# Calculate the cumulative returns and plot them
cumulative_returns = (1 + returns).cumprod()
fig_cum = px.line(cumulative_returns, title="Cumaltive Returns (Growth of $1 Investment)", labels={"value": "Invesetment Value ($)", "Date": "Month Year"})
fig_cum.show()
```

From the figure, WMT had the highest cumulative returns around $1.2 and the SPY had the lowest around $1. There was a crash in the stock market in 2022, but WMT held strong. This tells us that stock such as WMT, a place where people will shop regardless of the state economy, surive crashes while stock like AAPL suffer.

```python
# Caclulate the correlation and plot it as a matrix
correlation = returns.corr()
fig_corr = px.imshow(correlation, text_auto=True, title="Stock Correlation Matrix", color_continuous_scale="RdBu")
fig_corr.show()
```

From the matrix, we can see that SPY and AAPL have a strong correlation of 0.8465785, which tells us that Apple will most likely crash if the market crashes. WMT and SPY have a weak correlation of 0.3818974, which means Walmart does not usually follow the market, and a crash in the market would likely not affect Walmart.
